```python
%sql
-- Which wealth tier has the best portfolio returns & lowest churn?
SELECT 
    WealthTier,
    total_customers,
    ROUND(avg_net_assets, 0)            AS avg_net_assets,
    ROUND(avg_portfolio_return * 100, 2) AS avg_return_pct,
    ROUND(avg_investments_held, 1)       AS avg_investments,
    ROUND(avg_diversification * 100, 1)  AS avg_diversification_pct,
    total_churned,
    ROUND((total_churned / total_customers) * 100, 1) AS churn_rate_pct
FROM gold_wealth_tiers
ORDER BY avg_net_assets DESC
```

```python
%sql
-- Which age group invests most diversely and earns the best returns?
SELECT
    AgeGroup,
    total_customers,
    ROUND(avg_net_assets, 0)        AS avg_net_assets,
    ROUND(avg_return * 100, 2)      AS avg_return_pct,
    ROUND(avg_investments, 1)       AS avg_investments_held,
    ROUND(avg_risk_profile, 2)      AS avg_risk_profile,
    total_churned
FROM gold_age_groups
ORDER BY avg_return_pct DESC
```

```python
%sql
-- How do customers compare across France, Spain and Germany?
SELECT
    Country,
    WealthTier,
    total_customers,
    ROUND(avg_net_assets, 0)     AS avg_net_assets,
    ROUND(avg_return * 100, 2)   AS avg_return_pct,
    ROUND(avg_credit_score, 0)   AS avg_credit_score,
    ROUND(avg_products_held, 1)  AS avg_products_held,
    total_churned,
    ROUND((total_churned / total_customers) * 100, 1) AS churn_rate_pct
FROM gold_countries
ORDER BY Country, avg_net_assets DESC
```

```python
%sql
-- Do customers with better credit scores earn better portfolio returns?
SELECT
    CreditBand,
    COUNT(*) AS total_customers,
    ROUND(AVG(Net_Assets), 0)        AS avg_net_assets,
    ROUND(AVG(Portfolio_Return) * 100, 2) AS avg_return_pct,
    ROUND(AVG(CreditScore), 0)       AS avg_credit_score,
    ROUND(AVG(InvestmentCount), 1)   AS avg_investments,
    SUM(Churn)                        AS total_churned
FROM wealth_silver
GROUP BY CreditBand
ORDER BY avg_credit_score DESC
```

```python
%sql
-- Who are the most valuable customers by net assets?
SELECT
    CustomerID,
    Country,
    Age,
    AgeGroup,
    WealthTier,
    CreditBand,
    ROUND(Net_Assets, 0)         AS net_assets,
    ROUND(Portfolio_Return * 100, 2) AS return_pct,
    InvestmentCount,
    NumProducts,
    Churn
FROM wealth_silver
ORDER BY Net_Assets DESC
LIMIT 10
```

```python
%sql
-- Which customer profile churns the most?
SELECT
    WealthTier,
    AgeGroup,
    CreditBand,
    Country,
    COUNT(*)   AS total_customers,
    SUM(Churn) AS churned,
    ROUND((SUM(Churn) / COUNT(*)) * 100, 1) AS churn_rate_pct,
    ROUND(AVG(Net_Assets), 0) AS avg_net_assets
FROM wealth_silver
GROUP BY WealthTier, AgeGroup, CreditBand, Country
HAVING COUNT(*) > 10
ORDER BY churn_rate_pct DESC
LIMIT 15
```

```python
%sql
-- Which investment products are most popular across wealth tiers?
SELECT
    WealthTier,
    ROUND(AVG(EmergingMarketFund) * 100, 1) AS pct_emerging_market,
    ROUND(AVG(RealEstate) * 100, 1)         AS pct_real_estate,
    ROUND(AVG(PrivateEquity) * 100, 1)      AS pct_private_equity,
    ROUND(AVG(GovtBonds) * 100, 1)          AS pct_govt_bonds,
    ROUND(AVG(CorpBonds) * 100, 1)          AS pct_corp_bonds,
    ROUND(AVG(ETF_Tech) * 100, 1)           AS pct_etf_tech,
    ROUND(AVG(ETF_Health) * 100, 1)         AS pct_etf_health,
    ROUND(AVG(ETF_Med) * 100, 1)            AS pct_etf_med
FROM wealth_silver
GROUP BY WealthTier
ORDER BY WealthTier
```
