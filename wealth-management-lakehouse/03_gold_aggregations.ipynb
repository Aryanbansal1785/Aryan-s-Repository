```python
# ── GOLD LAYER: Wealth Management Analytics
from pyspark.sql import functions as F

df = spark.read.table("wealth_silver")

# ── Gold Table 1: Wealth tier performance summary
# Question: How do portfolio returns differ across wealth tiers?
gold_wealth_tiers = df.groupBy("WealthTier") \
    .agg(
        F.count("*").alias("total_customers"),
        F.avg("Net_Assets").alias("avg_net_assets"),
        F.avg("Portfolio_Return").alias("avg_portfolio_return"),
        F.avg("CreditScore").alias("avg_credit_score"),
        F.avg("InvestmentCount").alias("avg_investments_held"),
        F.avg("EstimatedSalary").alias("avg_salary"),
        F.sum("Churn").alias("total_churned"),
        F.avg("Diversification").alias("avg_diversification")
    ) \
    .orderBy(F.desc("avg_net_assets"))

gold_wealth_tiers.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_wealth_tiers")

print("✅ Gold Table 1: Wealth Tiers written")
display(gold_wealth_tiers)


# ── Gold Table 2: Age group investment behaviour
# Question: Which age groups invest most diversely and earn best returns?
gold_age_groups = df.groupBy("AgeGroup") \
    .agg(
        F.count("*").alias("total_customers"),
        F.avg("Net_Assets").alias("avg_net_assets"),
        F.avg("Portfolio_Return").alias("avg_return"),
        F.avg("InvestmentCount").alias("avg_investments"),
        F.avg("Risk_Profile").alias("avg_risk_profile"),
        F.avg("ForeignAssets").alias("avg_foreign_assets"),
        F.sum("Churn").alias("total_churned")
    ) \
    .orderBy("AgeGroup")

gold_age_groups.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_age_groups")

print("✅ Gold Table 2: Age Groups written")
display(gold_age_groups)


# ── Gold Table 3: Country level portfolio analysis
# Question: How do customers differ across France, Spain, Germany?
gold_countries = df.groupBy("Country", "WealthTier") \
    .agg(
        F.count("*").alias("total_customers"),
        F.avg("Net_Assets").alias("avg_net_assets"),
        F.avg("Portfolio_Return").alias("avg_return"),
        F.avg("CreditScore").alias("avg_credit_score"),
        F.avg("NumProducts").alias("avg_products_held"),
        F.sum("Churn").alias("total_churned"),
        F.avg("EstimatedSalary").alias("avg_salary")
    ) \
    .orderBy("Country", F.desc("avg_net_assets"))

gold_countries.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_countries")

print("✅ Gold Table 3: Countries written")
display(gold_countries)
```
